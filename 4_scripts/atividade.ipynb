```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
navegador = webdriver.Chrome()
```

```python
navegador.get('https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_PIB_nominal')
```

```python
pais = navegador.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[1]/td[1]/a').text
regiao = navegador.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[1]/td[2]/a').text
fmi_estimativa = navegador.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[1]/td[3]').text
fmi_ano = navegador.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[1]/td[4]').text
bm_estimativa = navegador.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[1]/td[5]').text

```

```python
print(pais,regiao,fmi_estimativa,fmi_ano,bm_estimativa)
```

```python
lista_vazia_pais = []
for i in range (1,101):
    pais = navegador.find_element(By.XPATH, '//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr['+str(i)+']/td[1]/a').text
    lista_vazia_pais.append(pais)
```

```python
df = pd.DataFrame(lista_vazia_pais, columns=['pais'])
```
